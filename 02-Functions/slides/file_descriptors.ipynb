# File descriptors

В Unix всё - файл

Every I/O operation uses a file descriptor

Использую файловые дескрипторы в:

1. Чтение и запись файлов
2. Директории
3. Девайсы
4. Коммуникация между процессами
5. Сетевая коммуникация
6. I/O multiplexing
7. Мониторинг файловой системы
8. ...

Почти как билеты, получив его от ядра, вы каждый раз при желании сделать что-нибудь с файлом предъявляете его.


Стандартные числа

0 - standard input

1 - standard output

2 - standard error output



-1: error - Python raises an exception

# Сколько здесь файловых дескрипторов?

```python
with open("example.txt") as f:
    print(f.read())
```

1 - running process

2 - open creates a file descriptor

3 - print writes to another file descriptor

### Можем использовать дескриптор как ссылку на файл
когда он уже открыт

```python
with open("example.txt") as f:
    stat = os.stat(f.fileno())
    print(stat)
    print(f.fileno())
    
    os.chmod(f.fileno(), 777)
```
